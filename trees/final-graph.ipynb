```python
import pandas as pd
import matplotlib.pyplot as plt


''' for ordered list '''
df_insertion_list = pd.read_csv('./csv_orderedlist/insertion.csv', header=0)
df_all_insertion = pd.read_csv('./csv_orderedlist/insertion_all.csv', header=0)
df_search_list = pd.read_csv('./csv_orderedlist/search.csv', header=0)
df_all_search = pd.read_csv('./csv_orderedlist/search_all.csv', header=0)
df_delete_list = pd.read_csv('./csv_orderedlist/delete.csv', header=0)
# df_cut_se = df_search_list

# ''' BSTcsvs'''
df_delete_BST = pd.read_csv('./csv_bst/delete.csv', header=0)
df_all_search_BST = pd.read_csv('./csv_bst/search_all.csv', header=0)
df_all_insertion_BST = pd.read_csv('./csv_bst/insertion_all.csv', header=0)

'''BBSTcsvs'''
df_insertion_BBST = pd.read_csv('./csv_bbst/insertion.csv', header=0)
df_search_BBST = pd.read_csv('./csv_bbst/search.csv', header=0)
df_delete_BBST = pd.read_csv('./csv_bbst/delete.csv', header=0)
```

```python
''' ************            ordered list          ************  '''
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white
#df_insertion_BST.loc[(df_insertion_BST['n'] < max_el)]
max_el=30000
plt.plot(df_all_insertion.loc[(df_all_insertion['n'] < max_el)]['n'], df_all_insertion.loc[(df_all_insertion['n'] < max_el)]['time'], label="insert all", color='purple')
plt.plot(df_all_search.loc[(df_all_search['n'] < max_el)]['n'], df_all_search.loc[(df_all_search['n'] < max_el)]['time'], label="search all", color='red')
plt.plot(df_delete_list['n'], df_delete_list['time'], label="delete all")
plt.title("Ordered list")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
''' ************            BST          ************  '''
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white
max_el=950000
plt.plot(df_all_insertion_BST.loc[(df_all_insertion_BST['n'] < max_el)]['n'], df_all_insertion_BST.loc[(df_all_insertion_BST['n'] < max_el)]['time'], label="insert all")
plt.plot(df_all_search_BST.loc[(df_all_search_BST['n'] < max_el)]['n'], df_all_search_BST.loc[(df_all_search_BST['n'] < max_el)]['time'], label="search all")
plt.plot(df_delete_BST.loc[(df_delete_BST['n'] < max_el)]['n'], df_delete_BST.loc[(df_delete_BST['n'] < max_el)]['time'], label="delete all")
#plt.plot(df_delete_BST['n'], df_delete_BST['time'], label="delete")
plt.title("BST")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
''' ************            BBST          ************  '''
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white
max_el=160001
plt.plot(df_insertion_BBST.loc[(df_insertion_BBST['n'] < max_el)]['n'], df_insertion_BBST.loc[(df_insertion_BBST['n'] < max_el)]['time'], label="insertion")
plt.plot(df_search_BBST.loc[(df_search_BBST['n'] < max_el)]['n'], df_search_BBST.loc[(df_search_BBST['n'] < max_el)]['time'], label="search")
plt.plot(df_delete_BBST.loc[(df_delete_BBST['n'] < max_el)]['n'], df_delete_BBST.loc[(df_delete_BBST['n'] < max_el)]['time'], label="delete")
#plt.plot(df_delete_BBST['n'], df_delete_BBST['time'], label="delete")
plt.title("BBST")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```
