```python
import pandas as pd
import matplotlib.pyplot as plt

df_withBS = pd.read_csv('A.csv', header=0)
df_withoutBS = pd.read_csv('B.csv', header=0)


names = ["HS", "BS", "ShS", "CS"]
with_dataframes = [df_withBS.loc[df_withBS['type'] == "HS"], df_withBS.loc[df_withBS['type'] == "BS"], df_withBS.loc[df_withBS['type'] == "ShS"], df_withBS.loc[df_withBS['type'] == "CS"]]
without_dataframes = [df_withoutBS.loc[df_withoutBS['type'] == "HS"], df_withoutBS.loc[df_withoutBS['type'] == "ShS"], df_withoutBS.loc[df_withoutBS['type'] == "CS"]]
data_types = ["random", "constant", "increasing", "decreasing", "A-shaped", "V-shaped"]
names_without = ["HS", "ShS", "CS"]

n=6
```

```python
# With BS
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white
for i in range(0, 4):
    plt.plot(with_dataframes[i]['n'], with_dataframes[i]['time'], label=names[i])
plt.title("HS, BS, ShS and CS")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
# Without BS
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white
for i in range(3):
    plt.plot(without_dataframes[i]['n'], without_dataframes[i]['time'], label=names_without[i])
plt.title("HS, ShS and CS")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```
