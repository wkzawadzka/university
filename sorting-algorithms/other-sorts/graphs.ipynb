```python
import pandas as pd
import matplotlib.pyplot as plt

''' complete following ~~~'''
df_mshs = pd.read_csv('withoutqs.csv', header=0)
df_rcid = pd.read_csv('withoutVA.csv', header=0)
df_650 = pd.read_csv('output.csv', header=0)
''' ____________________________________'''

algo = ["QS", "MS", "HS"]
mshs_dataframes = [df_mshs.loc[df_mshs['type'] == "MS"], df_mshs.loc[df_mshs['type'] == "HS"]]
rcid_dataframes = [df_rcid.loc[df_rcid['type'] == "QS"], df_rcid.loc[df_rcid['type'] == "MS"], df_rcid.loc[df_rcid['type'] == "HS"]]
dataframes_650 = [df_650.loc[df_650['type'] == "QS"], df_650.loc[df_650['type'] == "MS"], df_650.loc[df_650['type'] == "HS"]]
data_types = ["random", "constant", "increasing", "decreasing", "A-shaped", "V-shaped"]

n=6
```

```python
# Quicksort all 60,000
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white
for i in range(0, n):
    plt.plot(dataframes_650[0].loc[dataframes_650[0]['data'] == data_types[i]]['n'], dataframes_650[0].loc[dataframes_650[0]['data'] == data_types[i]]['time'], label=data_types[i])
plt.title("Quicksort")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
# Mergesort all 5M
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white
for i in range(0, n):
    x = mshs_dataframes[0].loc[df_mshs['data'] == data_types[i]]
    y = mshs_dataframes[0].loc[df_mshs['data'] == data_types[i]]
    plt.plot(x['n'], y['time'], label=data_types[i])
plt.title("Mergesort")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
# Heapsort all 5M
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white
for i in range(0, n):
    x= mshs_dataframes[1].loc[df_mshs['data'] == data_types[i]]
    y = mshs_dataframes[1].loc[df_mshs['data'] == data_types[i]]
    plt.plot(x['n'], y['time'], label=data_types[i])
plt.title("Heapsort")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
# Random and Constant 5M
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white

for algorithm in range(3):
    for datatype in range(2):
        x = rcid_dataframes[algorithm].loc[df_rcid['data'] == data_types[datatype]]
        y = rcid_dataframes[algorithm].loc[df_rcid['data'] == data_types[datatype]]
        plt.plot(x['n'], y['time'], label= algo[algorithm] + " " + data_types[datatype])

plt.title("Random and constant data type")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
# Increasing and decreasing 5M
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white

for algorithm in range(3):
    for datatype in range(2,4):
        x = rcid_dataframes[algorithm].loc[df_rcid['data'] == data_types[datatype]]
        y = rcid_dataframes[algorithm].loc[df_rcid['data'] == data_types[datatype]]
        plt.plot(x['n'], y['time'], label= algo[algorithm] + " " + data_types[datatype])

plt.title("Inreasing and decreasing data type")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
# A and V
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white

for algorithm in range(1,3):
    for datatype in range(4,6):
        x = dataframes_650[algorithm].loc[df_650['data'] == data_types[datatype]]
        y = dataframes_650[algorithm].loc[df_650['data'] == data_types[datatype]]
        plt.plot(x['n'], y['time'], label= algo[algorithm] + " " + data_types[datatype])

plt.title("A-shaped and v-shaped data type\nwith worstcase for QS")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```

```python
# A and V
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white') # make transparent background white

for algorithm in range(0,1):
    for datatype in range(4,6):
        x = dataframes_650[algorithm].loc[df_650['data'] == data_types[datatype]]
        y = dataframes_650[algorithm].loc[df_650['data'] == data_types[datatype]]
        plt.plot(x['n'], y['time'], label= algo[algorithm] + " " + data_types[datatype])

plt.title("Worstcase for QS")
plt.xlabel('number of elements')
plt.ylabel('time [s]')
plt.legend()
plt.show()
```
